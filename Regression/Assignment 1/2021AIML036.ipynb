The following dataset of size n = 31 provides employee-wise salary. The variables are y = salary of an individual employee and x = years of experience of each individual employee . 
Find the intercept, coefficient value and error term for this regression problem
in Python. How would you convey (interpret) these results in a business
context?		

```python
### here we have to fing y = w0 + w1.x where x represents YearsExperience and Y is target salary
### w0 is intercept and w1 is coefficient , e is error term.
### error function = 1/2 * Sum ((w0 + w1.x) - y)^2)
### when we take derivative of error function and eqaute it to 0, we arrive at below 2 equation

### From buisness perspective A positive coefficent value dertermines that dependent variable (salery) is increasing with 
### increase in the independent variable (YearsExperience)

### RMSE value indicates the differnce between pedicted valued by model and actual value of salery.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r"C:\Users\visoni\Desktop\bits pilani\Assignment 1\Salary_Data.csv")
x = data['YearsExperience']
y = data['Salary']
print(data.head())

N = len(x)
x_mean = x.mean()  #calculate mean of yearsExperience 
y_mean = y.mean()  #calculate mean of salery of employee

w1 = ((x - x_mean) * (y - y_mean)).sum()/((x - x_mean)**2).sum() 
w0 = y_mean - (w1*x_mean)

reg_line = 'y = {} + {}x'.format(round(w0,3), round(w1,3))

print('Intercept value is : ', w0)
print('coefficient value is : ', w1)
print('Regression Line: ', reg_line)

plt.title("YearsExperince VS salery model")
plt.xlabel('Years of Experience', fontsize=15)
plt.ylabel('Salary', fontsize=15)
plt.plot(x, w0 + w1*x, c = 'r', linewidth=5, alpha=.5, solid_capstyle='round')
plt.scatter(x, y, s=300, linewidths=1, edgecolor='black')

RMSE = (np.sqrt(np.square(np.subtract(y, (w0 + w1*x))).mean()))
RMSE_normalized = (np.sqrt(np.square(np.subtract(y, (w0 + w1*x))).mean()))/y_mean
print("root mean square error is : ", RMSE )
print("normalised root mean square error is : ", RMSE_normalized )
```
